```python
import pandas as pd
import re
```

```python
# Load the dataset
items = pd.read_csv("./kaggle_data/items.csv")
display(items.head())
```

```python
items['Title'] = items['Title'].str.rstrip(' /')
items
```

```python
def clean_author(author):
    if pd.isna(author):
        return None

    author = author.strip()

    author = re.sub(r'\([^)]*\)', '', author)

    author = re.sub(r'\d{2,4}[.-]*\d*[.-]*\d*', '', author)

    author = re.sub(r'[.]+', '', author)

    if ',' in author:
        parts = [p.strip() for p in author.split(',')]
        if len(parts) >= 2:
            author = f"{parts[1]} {parts[0]}"
            if len(parts) > 2:
                author += ' ' + ' '.join(parts[2:])

    author = re.sub(r'\s+', ' ', author).strip()

    return author
```

```python
items['author_clean'] = items['Author'].apply(clean_author)
```
